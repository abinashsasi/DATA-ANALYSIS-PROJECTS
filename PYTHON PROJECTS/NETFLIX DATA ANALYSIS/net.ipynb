```python
def get_top_actors(df_subset, column='cast', top_n=25):
    df_subset = df_subset[df_subset[column].notna()]
    all_actors = ", ".join(df_subset[column].values).split(", ")
    actor_counts = Counter(all_actors)
    top_actors = actor_counts.most_common(top_n)
    
    labels = [actor for actor, count in top_actors][::-1]
    values = [count for actor, count in top_actors][::-1]
    return labels, values
```

```python
def create_actor_dropdown(filter_column, df, top_n=None, x=0, y=1.2):
    options = df[filter_column].dropna()

    if filter_column == 'country':
      options = options.str.split(', ').explode()
      options = options.str.strip(", ").dropna()

    values = [val for val, _ in Counter(options).most_common(top_n)]  # top 8 for dropdown

    buttons = []

    # All Button
    labels, vals = get_top_actors(df, 'cast', top_n)
    buttons.append(dict(
        label='All',
        method='restyle',
        args=[{'x': [vals], 'y': [labels], 'text': [vals]}]
    ))

    for val in values:
        if filter_column == 'country':
            filtered = df[df[filter_column].str.contains(val, na=False)]
        else:
            filtered = df[df[filter_column] == val]

        labels, vals = get_top_actors(filtered, 'cast', top_n)

        buttons.append(dict(
            label=str(val),
            method='restyle',
            args=[{'x': [vals], 'y': [labels], 'text': [vals]}]
        ))

    dropdown = dict(
        buttons=buttons,
        direction="down",
        showactive=True,
        x=x,
        xanchor="left",
        y=y,
        yanchor="top",
        pad={"r": 10, "t": 10},
        bgcolor="#eeeeee",
        font=dict(size=12),
    )
    return dropdown
```

```python
def get_column_data(df_subset, column_name):
    value_counts = df_subset[column_name].value_counts().reset_index()
    value_counts.columns = [column_name, 'count']
    return value_counts
 

# Updated dropdown function to work with any column
def create_dropdown(filter_column, df, chart_column='rating', top_n=None, x=0, y=1.2, show_all=False):
    options = df[filter_column].dropna()

    # Handle multiple countries (if filter_column is 'country')
    if filter_column == 'country':
      options = options.str.split(', ').explode()
      options = options.str.strip(", ").dropna()

    unique_values = options.unique()
    if top_n and not show_all:
        values = [val for val, _ in Counter(options).most_common(top_n)]
    else:
        values = sorted(unique_values)

    buttons = []

    # All option (no filter)
    all_data = get_column_data(df, chart_column,)
    buttons.append(
        dict(
            label='All',
            method="restyle",
            args=[{
                "values": [all_data['count']],
                "labels": [all_data[chart_column]],
                "pull": [0.06] * len(all_data)
            }]
        )
    )

    for val in values:
        if filter_column == 'country':
            filtered = df[df[filter_column].str.contains(val, na=False)]
        else:
            filtered = df[df[filter_column] == val]

        data = get_column_data(filtered, chart_column)

        buttons.append(
            dict(
                label=str(val),
                method="restyle",
                args=[{
                    "values": [data['count']],
                    "labels": [data[chart_column]],
                    "pull": [0.06]*len(data)
                }]
            )
        )

    dropdown = dict(
        buttons=buttons,
        direction="down",
        showactive=True,
        x=x,
        xanchor="left",
        y=y,
        yanchor="top",
        pad={"r": 10, "t": 10},
        bgcolor="#eeeeee",
        font=dict(size=12),
    )

    return dropdown
```
